```python
df = pd.read_csv("training_data_normal.tsv", sep='\t', header=0)
df.head()
```

```python
x = df.iloc[:,:-1]
y = df['age']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2,random_state=3303421)

```

```python
selector = SelectKBest(f_regression,k=5890)

selector.fit(x_train,y_train)
x_train = selector.transform(x_train)
x_test = selector.transform(x_test)

x_train.shape

reg = ElasticNet(alpha=0.00001,normalize=False)
reg.fit(x_train,y_train)
preds = reg.predict(x_test)
r2_score(y_test,preds)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
k = [5880,5890,5895,5900]
alpha = [0.00001,0.0001,0.001]
normalise = [True, False]

import warnings
warnings.filterwarnings("ignore")

for val in k:
    for a in alpha:
        for n in normalise:
            selector = SelectKBest(f_regression,k=val)
            selector.fit(x_train,y_train)
            x_train_new = selector.transform(x_train)
            x_test_new = selector.transform(x_test)
            reg = ElasticNet(alpha=a,normalize=n,random_state=33032141)
            reg.fit(x_train_new,y_train)
            preds = reg.predict(x_test_new)
            score = r2_score(y_test,preds)
            print(f"score =, {score}. k=,{val}. alpha =,{a}. Norm = {n}")
```
